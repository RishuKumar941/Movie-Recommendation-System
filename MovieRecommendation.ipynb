Importing the Dependencies


```python
import numpy as np
import pandas as pd
import difflib  #used to make the comparision between the two
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and Pre Processing

```python
#loading the data from the csv file to the pandas dataframe
movies_data=pd.read_csv('/content/drive/MyDrive/movies.csv')
movies_data
```

```python
#Printing the first 5 data
movies_data.head()
```

```python
#to find the number of rows and columns in the data frame
movies_data.shape
```

```python
#selecting the Relevant features from the data frame
selected_feature=['genres','keywords','tagline','cast','director']
print(selected_feature)
```

```python
#replacing the null values with the null string
for feature in selected_feature:
  movies_data[feature]=movies_data[feature].fillna('')
```

```python
#combining all the features
combined_features = movies_data['genres']+''+movies_data['keywords']+''+movies_data['tagline']+''+movies_data['cast']+''+movies_data['director']
```

```python
print(combined_features)
```

```python
#converting the text data to feature vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
#getting the similarity scores using the cosine similarity
similarity=cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

Getting a movie name from the user

```python
#getting the movie anme from the user
movie_name=input('Enter the favourite movie name :')
```

```python
#creating a list with all the movies name given in the dataset
list_of_all_titles=movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
#finding the close match given by the user for an movie name
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
#finding the index of the movies with the title
index_of_the_movies=movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movies)
```

```python
#getting an list of all the Similar movies

similarity_score = list(enumerate(similarity[index_of_the_movies]))
print(similarity_score)
```

Movie Recommendation System


```python
movie_name = input(' Enter your favourite movie name : ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movies = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movies]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
